### Generation des mini-challenges

```python
prompt_structure_mini_challenges="""

Agie comme un generateur des mini-challenges des jeux pedagogiques, 
je vous donne une instruction qui represente un mecanique didactique (MD) et un mini-scenario (MS) 
et il faut generer des mini-challenge differents associer a cette instruction: 

Exemple:

Instruction :

    Le joueur se rappelle la table de multiplication de 4 (MD) 
    et doit surmonter un défi qui bloque son avancement en donnant le résultat de la multiplication d'un nombre par 4 (MS).

Mini-challenges:

MC1 : Le joueur se trouve face à un adversaire redoutable qui lance des cartes maléfiques. Chaque carte affiche un nombre mystérieux. 
Pour progresser, le joueur doit utiliser sa connaissance de la table de multiplication de 4 et calculer rapidement le produit de 4 
par le nombre présent sur la carte. Si sa réponse est correcte, il remporte des points et peut continuer son chemin. 
En cas d'échec, l'adversaire gagne en puissance et le joueur risque de perdre des avantages précieux. 

MC2 : Le joueur se tient devant une imposante porte verrouillée, arborant un nombre énigmatique, comme 7. Pour franchir cette porte et poursuivre 
son périple, le joueur doit mettre en pratique sa connaissance des tables de multiplication. En l'occurrence, il doit résoudre le calcul 7 multiplié par 4.
Si sa réponse est juste, la porte s'ouvrira, révélant de nouvelles possibilités. En revanche, en cas d'erreur, la porte demeurera fermée, 
empêchant toute progression jusqu'à ce que le joueur trouve la solution. 

MC3 : Une rivière tumultueuse barre la route du joueur, dont le seul espoir de traversée réside dans la réparation d'un pont effondré. 
Trois piliers émergent des flots, mais pour faire apparaître le pont salvateur, le joueur doit mettre à profit ses connaissances mathématiques.
En répondant correctement à la multiplication de 4 par 3, il déclenche le processus magique qui reconstruit le pont et lui permet de continuer 
son périple. Toutefois, une erreur le condamnerait à rester bloqué de l'autre côté, confronté aux dangers de l'inconnu.

Resultat sous forme des paragraphes.
Generer des mini-challenges pour cette instruction: {}

"""
```

```python
# Exemple
full_prompt = prompt_structure_mini_challenges.format("Le joueur doit identifier les parties du corps humain (MD) en associant chaque partie à son nom en anglais dans un jeu de memory (MS).")
result = gpt(full_prompt)
print(result)
```

```python
prompt_structure_2="""

Voici un exemple pour analysé:

un mecanique dedactique: exemple Le joueur se rappelle la table de multiplication de 4.
un mini-scénario didactique qui va utiliser cette mécanique didactique: exemple présenter un défi qui bloque l’avancement du joueur.Le défi porte un nombre. Le joueur doit répondre au défi en donnant le résultat de multiplication du nombre par 4. *
pour ce mini scenario on peut generer plusieurs mini challenges comme : exemple le joueur est devant un opposant qui sort des cartes maléfiques. Chaque carte comporte un nombre. Le joueur répond à la multiplication de 4 par le nombre de la carte pour gagner des points.
Ou le joueur est devant 1 porte avec un nombre dessus (exemple 7). Le joueur doit donner le résultat de 7x4 pour ouvrir la porte.
 
On construirant un dataset sous forme Q&A voici la qestion convenable a ce MD+MS

Quels mini-challenges pourraient être proposés dans un scénario où le joueur doit se rappeler la table de multiplication de 4 pour avancer en répondant à des défis numérotés?

En inspirant de exemple faire :

Acter comme un concepteur des jeux pedagogique, Tu specialiste de generation des mecaniques dedactiques des jeux. en ce basant sur le theme suivant:

  {}
  
Structure de resultat doit etre: MD: MS: CH: Q:
"""
```

```python
import re

patterns = {
    "MD": r"MD:\s*(.*?)\s*MS:",
    "MS": r"MS:\s*(.*?)\s*CH:",
    "CH": r"CH:\s*(.*?)\s*Q:",
    "Q": r"Q:\s*(.*)"
}
```

```python
# Créer le DataFrame
mecaniques = []

# Parcourir chaque ligne du DataFrame
for row in random_themes:

    # Définir le prompt complet avec la description du jeu
    full_prompt = prompt_structure_2.format(row)
    
    result = gpt(full_prompt)

    print(result)
    print("=============")
    
    # Dictionnaire pour stocker les résultats
    extracted = {}

    # Extraire chaque partie
    for key, pattern in patterns.items():
        match = re.search(pattern, result, re.DOTALL)
        if match:
            extracted[key] = match.group(1).strip()

    if extracted['Q'] is not None and extracted['CH'] is not None:    
        mecaniques.append([extracted['Q'],extracted['CH']])
        # Dictionnaire pour stocker les résultats
        extracted = {}


# Afficher le DataFrame mis à jour avec les histoires générées
print(mecaniques)
```

```python
# Créer DataFrame
df = pd.DataFrame(mecaniques, columns=['Question', 'Answer'])

# Afficher DataFrame
df
```
