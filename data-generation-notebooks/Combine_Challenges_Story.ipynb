```python
def gpt(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        temperature = 0,
        max_tokens=1500,
        messages=[
            {
                "role": "user",
                "content": prompt,
            }
        ],
        )
    reply_content = response.choices[0].message.content.strip()
    return reply_content
```

### Generation version final dataset combiné

```python
prompt_structure= """
Agie comme generateur des game stories avec des challenge:

Voici un conte de fées:

{}

Voici des mini challenges :

{}

Resultat contient obligatoirement:
paragraphe 1 : une idée sur cadre spatio-temporel
paragraphe 2 : une description des personnage (physique,morale,caracteristiques,états...)
ensuite:
La situation initiale
Un évènement déclencheur et processus narratif (amorce et déroulement de la quête) souvent accompagné d’évènements perturbateurs
3 péripéties des héros (paragraphes chaqu'un prend en charge un MINI-CHALLENGE lié avec histiore du 
joueur hero deja mentionner pour atteindre objectif final du game .)
Le dénouement est l’ultime quête du héros
La situation finale résulte de cette ultime quête

sous forme paragraphes sans titres, sans mentionner les nom de parties dans l'histoire (La situation initiale,premiere péripéties...)
les noms des personnages dans le text doivent etre entre les caractéres ""

(ne pas mentionner les titres de structure)
"""
```

```python
mini_challenges = df_challenges.iloc[0, 1]
story = df_stories.iloc[0, 3]

# Définir le prompt complet avec la description du jeu
full_prompt = prompt_structure.format(story,mini_challenges)

# Extraire l'histoire générée
generated_story = gpt(full_prompt)

print(generated_story)
print("----------------------------------------------")

df_challenges.iloc[0, 1] = df_challenges.iloc[0, 1] +"\n\n"+ generated_story
```

```python
# Parcourir chaque ligne du DataFrame
for index,row in df_challenges.iterrows():
    
    mini_challenges = row['OUTPUT']
    story = df_stories.iloc[index, 3]
    
    # Définir le prompt complet avec la description du jeu
    full_prompt = prompt_structure.format(story,mini_challenges)
    
    # Extraire l'histoire générée
    generated_story = gpt(full_prompt)
    
    print(generated_story)
    print("----------------------------------------------")
    
    row['OUTPUT'] = row['OUTPUT'] +"\n\n"+ generated_story
```
